# integreating MLFlow with GenAI applications 

```python
import mlflow
import os 
from langchain.chat_models import AzureChatOpenAI
from openai import AzureOpenAI
from langchain.schema import SystemMessage, HumanMessage
from dotenv import load_dotenv

load_dotenv()
```

```python
llm=AzureChatOpenAI(
    openai_api_version=os.getenv('AZURE_OpenAI_API_VERSION'),
    azure_deployment=os.getenv('AZURE_OPENAI_DEPLOYMENT_MODEL'),
    model_name=os.getenv('AZURE_OPENAI_DEPLOYMENT_MODEL'),
    api_key=os.getenv('GRAPHRAG_API_KEY')
)
```

```python
message=[
    SystemMessage(
        content=("you are a sarcastic batman")

    ),
    HumanMessage(content='why the bat suit?')
]

res=llm.invoke(message)
print(res.content)
```

```python
from openai import AzureOpenAI

client=AzureOpenAI(
    api_key=os.getenv('GRAPHRAG_API_KEY'),
    api_version='2023-05-15',
    azure_endpoint=os.getenv('AZURE_OPENAI_EMBEDDINGS_ENDPOINT')
)

response= client.embeddings.create(
    input='amruth pai thukaram',
    model=os.getenv('AZURE_OPENAI_EMBEDDING_MODEL_NAME')
)
print(response.model_dump_json(indent=2))
```
